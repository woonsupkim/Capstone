```python
import sys
sys.path.insert(1, 'C:/Users/Woon/Desktop/Columbia/Applied Analytics/Term3/Capstone/lib')

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

pd.set_option('display.max_columns', None) 
```

```python
df_demo = pd.read_csv('data\demo_data.csv')
df_demo = df_demo.drop(columns=['Unnamed: 0'])
df_demo
```

## Homedepot

```python
df_homedepot_price=df_demo[df_demo['price_m'] >= 0.7][df_demo[df_demo['price_m'] >= 0.7]['retailer'] == 'Homedepot']
tab1 = pd.crosstab(index=df_homedepot_price['sentiment'], columns='count')
100*tab1/tab1.sum()
print('\n')
df_homedepot_quality=df_demo[df_demo['quality_m'] >= 0.7][df_demo[df_demo['quality_m'] >= 0.7]['retailer'] == 'Homedepot']
tab2 = pd.crosstab(index=df_homedepot_quality['sentiment'], columns='count')
100*tab2/tab2.sum()
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(10,6)
plt.title('Homedepot Brands - Price');
hue_order = ['Positive', 'Negative', 'Neutral']
sns.countplot(y='brand', data=df_demo[df_demo['price_m'] >= 0.7][df_demo[df_demo['price_m'] >= 0.7]['retailer'] == 'Homedepot'], hue = 'sentiment');
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(10,6)
plt.title('Homedepot Brands - Quality');
hue_order = ['Positive', 'Negative', 'Neutral']
sns.countplot(y='brand', data=df_demo[df_demo['quality_m'] >= 0.7][df_demo[df_demo['quality_m'] >= 0.7]['retailer'] == 'Homedepot'], hue = 'sentiment');
```
